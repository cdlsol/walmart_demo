```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\carlo\walmart-sales\Walmart_Sales.csv')
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
display(df)
```

```python
df["year"] = df["Date"].dt.year
df["month"] = df["Date"].dt.to_period("M")
df["week"] = df["Date"].dt.to_period("W")

general_metrics = df.groupby("Store").agg(
    weekly_avg_sales=("Weekly_Sales", "mean"),
    monthly_total_sales=("Weekly_Sales", "sum"),
    yearly_total_sales=("Weekly_Sales", "sum")
).reset_index()
general_metrics["weekly_avg_sales"] = general_metrics["weekly_avg_sales"].round(2)
display(general_metrics)

```

```python
df2 = df.groupby("week").agg(
    weekly_sales=("Weekly_Sales", "sum")
).reset_index()
display(df2)
```

```python
df3 = df.groupby("month").agg(
    monthly_sales=("Weekly_Sales", "sum")
).reset_index()
display(df3)
```

```python
df4 = df.groupby("year").agg(
    yearly_sales=("Weekly_Sales", "sum")
).reset_index()
display(df4)
```
